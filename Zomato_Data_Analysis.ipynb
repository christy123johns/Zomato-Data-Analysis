```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
#trying to combine the country code of zomota data and country code data
```

```python
#to combine country codes of both data
# on try to ask which variable
# how ask in data which side it needed to added so we given left side
#call this data with a new name final_df
#also we can use concat code for merging
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(3)
```

```python
#to check data types
final_df.dtypes
```

```python
#going to explore data
final_df.columns
```

```python
final_df.Country.value_counts()
#zomata is mostly available in India, most of the peoples use zomato in india
#canada use less
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart- Top 3 country(to see top 3 country we use [:3] in code)
#x axis=country val
#autopct is auto percentage to give percentage values of each 
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.2f%%')

#observation : from this diagram zomata maximam records or transactions are from India, then USA and then UK
```

```python
final_df.columns
```

```python
#grouping these 3 variables and when we give size() we can see the result
#if we give reset_index the series data will convert into a dataframe.reseting data.
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
#renamed the table variable 0 into rating count
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings.head()

#observation : around 2148 data having zero ratings, when rating is between 4.5-4.9 it indicates the food delivered is excellent
#when rating 4.0-4.4 -very good rating, when rating between 3.5-3.9 good rating, 3.0-3.4 is average
# 2.5-2.9 rating is average, 2.0-2.4 is poor rating
#zero rating is given by many people
```

```python
#code for resizing graph
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
#bar chart code
sns.barplot(x='Aggregate rating',y='Rating count', data=ratings)

#observation : looks like a gaussian curve
```

```python
#hue will separate the dat using colour
sns.barplot(x='Aggregate rating',y='Rating count', hue='Rating color' , data=ratings)
```

```python
#to map colours according to our data we use palette
sns.barplot(x='Aggregate rating',y='Rating count', hue='Rating color' , data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green','darkgreen'])

#observation : 1. not rated count is very high,
# 2.maximum no of ratings are between 2.5 to3.4
```

```python
#count plot
sns.countplot(x='Rating color', data=ratings,palette=['blue', 'red', 'orange', 'yellow', 'green','darkgreen'])
```
